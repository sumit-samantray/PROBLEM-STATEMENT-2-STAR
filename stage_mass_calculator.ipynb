#### SOLVE FOR LAGRANGIAN MULTIPLIER (n):
|13.812 - c1*ln((c1*n - 1)/(c1*e*n) - c2*ln((c2*n - 1)/(c2*e*n)) - c3*ln((c3*n - 1)/(c3*e*n)| = 0

```python
import math
n = 0.651 # interpreted from the graph
c1 = 3 # characteristic velocity of phase 1
c2 = 3.5 # characterictic velocity of phase 2
c3 = 4 # characteristic velocity of stage 3
e = 0.15 # assumed structural ratio (usually between 0.1 and 0.2)
mpl = 1350 #payload mass kg
g = 9.81 #acceleration due to gravity m/s^2
n1 = (c1*n - 1)/(c1*n*e) # mass ratio of phase 1
n2 = (c2*n - 1)/(c2*n*e) # mass ratio of phase 2
n3 = (c3*n - 1)/(c3*n*e) # mass ratio of phase 3
m3 = (n3 - 1)*mpl/(1 - e*n3) # mass of third stage
m2 = (n2 - 1)*(mpl+m3)/(1 - e*n2) # mass of second stage
m1 = (n1 - 1)*(mpl+m3+m2)/(1 - e*n1) # mass of first stage
mo = m1 + m2 + m3 + mpl # total mass of the launch vehicle
me1 = e*m1 # structural mass of p - 1
mp1 = m1*(1-e) # propellant mass of p - 1
me2 = e*m2 # structural mass of p - 2
mp2 = m2*(1-e) # propellant mass of p - 2
me3 = e*m3 # structural mass of p - 3
mp3 = m3*(1-e) # propellant mass of p - 3
print("m1: ", round(m1, 3), "kg")
print("m2: ", round(m2, 3), "kg")
print("m3: ", round(m3, 3), "kg")
print("total mass: ", round(mo, 3), "kg")
```

```python
pc1 = 7 * 10**6 #pa combustion chamber pressure
mdot1 = 2578 #mass flow rate kg/s
at1 = mdot1*c1*1000/pc1 #throat area m^2
print("throat area of nozzle: ", round(at1, 3), "m^2")
thrust1 = mdot1*c1 # average thrust N
print("thrust: ", round(thrust1, 3), "kN")
b1 = (mo - m1*e)/mdot1 # burn time (s)
print("burn time: ", round(b1, 3), "s")
mo1 = 2.56*mp1/3.56 # oxidizer mass for phase 1
vo1 = mo1/1141 # volume of oxidizer for phase 1
mf1 = mp1/3.56 # mass of fuel for stage 1
vf1 = mf1/840 # volume of fuel for stage 1
print("volume of fuel tank: ", round(vf1, 3), "m^3")
print("volume of oxidiser tank: ", round(vo1, 3), "m^3")
```
